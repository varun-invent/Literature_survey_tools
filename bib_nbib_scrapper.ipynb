````python
in_file  = 'bib/scopus.bib'
# in_file  = 'bib/WebOf Science.bib'
entries = []
entry = {}


with open(in_file) as fd:
    for line in fd:
        print(line)
        if (match('^@', line.strip())):
            if entry != {}:
                entries.append(entry)
                entry = {}

            
        elif (match('title', line.lower().strip())):
            value, = findall('\{(.+)\}', line)
            entry["title"] = value
    
            
        elif (match('abstract', line.lower().strip())):
            value, = findall('\{(.+)\}', line)
            entry["abstract"] = value
            
````

# Now comparing the titles to find overlap 

1. Keep pubmed as the core DB
2. Check for other database, for each query title how much is the percent match with the words separated by '_', ' ' or '-'.
3. Keep the threshold at 60%

````python
cite_file_1 = 'pubmed12-359.csv'
cite_file_2 = 'scopus_bib.csv'
# cite_file_2 = 'webofscience_nbib.csv'
````

````python
file_1 = pd.read_csv(cite_file_1)
file_2 = pd.read_csv(cite_file_2)
````

````python
file_1.columns
````

````python
colums_ix_to_include = [1]
````

````python
file_1_columns = file_1.iloc[:,colums_ix_to_include]
file_2_columns = file_2.iloc[:,colums_ix_to_include]
````

````python
file_1_columns = file_1_columns.as_matrix().squeeze()
file_2_columns = file_2_columns.as_matrix().squeeze()
````

````python
len(file_1_columns),len(file_2_columns)
````

````python
len(set(file_1_columns)),len(set(file_2_columns))
````

````python
# Based on % of unordered words similarity

UNORDERED_WORD_SIMILARITY = False

if UNORDERED_WORD_SIMILARITY:
    common_index = []
    common_title = []
    word_count = 0
    matched = 0
    max_match_idx = 0

    for ix2, title_2 in tqdm(enumerate(file_2_columns)):
        match_percentage_old = 0
        max_match_idx = 0
        title_2_words = re.split("\s|_|:\s|\s:|:|-",title_2)
        for ix1, title_1 in enumerate(file_1_columns):
            for word in title_2_words:
                if word in title_1:
                    word_count += 1

            match_percentage =  word_count/len(title_2_words)
            if match_percentage > match_percentage_old:
                max_match_idx = ix1
                match_percentage_old = match_percentage

            word_count = 0
    #         print('Max match idx for ix2: %s is ix1: %s with match percent: %s'%(ix2,max_match_idx,match_percentage_old))

        if match_percentage > 0.20:
            common_title.append([file_2_columns[ix2],file_1_columns[max_match_idx]])
            common_index.append([ix2,max_match_idx])
            print('Matched: File2 Idx: %s with File1 Idx: %s with percentage = %s '%(ix2,max_match_idx,match_percentage))
            matched += 1

````

````python
# Based on ordered words similarity

ORDERED_WORD_SIMILARITY = False

if ORDERED_WORD_SIMILARITY:
    matched = 0
    match_length = 6
    common_index = []
    common_title = []
    for ix2, title_2 in tqdm(enumerate(file_2_columns)):
        title_2_words = re.split("\s|_|:\s|\s:|:|-|\s-|-\s",title_2)
        if '' in title_2_words:
            title_2_words.remove('')
        if ' ' in title_2_words:
            title_2_words.remove(' ')
        
        for ix1, title_1 in enumerate(file_1_columns):
            title_1_words = re.split("\s|_|:\s|\s:|:|-|\s-|-\s",title_1)
            if '' in title_1_words:
                title_1_words.remove('')
            if ' ' in title_2_words:
                title_2_words.remove(' ')
                
            if title_2_words[0:match_length]  == title_1_words[0:match_length]:
                common_title.append([file_2_columns[ix2],file_1_columns[ix1]])
                common_index.append([ix2,ix1])
                print('Matched: File2 Idx: %s with File1 Idx: %s'%(ix2,ix1))
                matched += 1
                    
````

````python
# removing the special characters and then setting them to lower case 
# Ref: https://stackoverflow.com/questions/5843518/remove-all-special-characters-punctuation-and-spaces-from-string

ORDERED_CHARACTERS_SIMILARITY = True

if ORDERED_CHARACTERS_SIMILARITY:
    matched = 0
    match_length = 40
    common_index = []
    common_title = []
    common_chars = []
    for ix2, title_2 in tqdm(enumerate(file_2_columns)):
        title_2_chars = (''.join(e for e in title_2 if e.isalnum())).lower()
        
        for ix1, title_1 in enumerate(file_1_columns):
            title_1_chars = (''.join(e for e in title_1 if e.isalnum())).lower()
                
            if title_2_chars[0:match_length]  == title_1_chars[0:match_length]:
#             if title_2_chars  == title_1_chars:
                common_title.append([file_2_columns[ix2],file_1_columns[ix1]])
                common_index.append([ix2,ix1])
                common_chars.append([title_2_chars,title_1_chars])
                print('Matched: File2 Idx: %s with File1 Idx: %s'%(ix2,ix1))
                matched += 1
                    
````
